```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')

plt.rcParams.update({
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
})
```

```python
df = pd.read_excel('kyoto_gases.xlsx').set_index('Scenario')
```

```python
fig, ax = plt.subplots()
df.T.plot(ax=ax)
ax.set_ylim(0, 11e4)
ax.set_xlim(2000, 2105)
ax.set_ylabel('Mt CO2-equivalent/yr')
ax.set_xlabel('Year')
plt.legend(fontsize='large', loc='best')
#plt.title('Global Kyoto Gas Emissions')
plt.tight_layout()
plt.savefig('../figs/results_kyoto.pdf')
```

```python
df = pd.read_excel('forcing.xlsx').set_index('Scenario')
harm = df[0:2].T
unharm = df[2:4].T
diff = df[4:].T * 100
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
harm.plot(ax=ax1, yerr=diff)
diff.plot(ax=ax2, legend=False, style='--')
ax1.set_ylim([0, 7])
ax2.set_ylim([-3.5, 3.5])
#ax2.grid(False)

ax1.set_ylabel('Radiative Forcing [W/m2]', fontsize='large')
ax1.set_xlabel('Year', fontsize='large')
ax2.set_ylabel('Relative Difference [%]', fontsize='large')
ax1.legend(fontsize='large', loc='best')

plt.savefig('../figs/results_forcing.pdf')
```

```python
# option 2
colors = plt.rcParams['axes.prop_cycle'].by_key()['color'][:2]
fig, ax1 = plt.subplots()
harm.plot(ax=ax1, color=colors)
unharm.plot(ax=ax1, color=colors, style='--')
ax1.set_ylim([0, None])
#ax2.grid(False)

ax1.set_ylabel('Radiative Forcing [W/m2]', fontsize='large')
ax1.set_xlabel('Year', fontsize='large')
ax1.legend(fontsize='large', loc='best')

plt.savefig('../figs/results_forcing.pdf')
```
